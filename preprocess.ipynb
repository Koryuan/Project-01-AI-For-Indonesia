```python
PATH_TRAIN = './train'
PATH_TEST = './test'
```

```python
if(os.path.exists(PATH_TRAIN)):
  shutil.rmtree(PATH_TRAIN)

if(os.path.exists(PATH_TEST)):
  shutil.rmtree(PATH_TEST)

os.makedirs(PATH_TRAIN)
os.makedirs(PATH_TEST)
```

```python
df = pd.read_csv('list_attribute.txt', sep='\s+', skiprows=1)
df.describe()
```

```python
file_images = os.listdir('images/')
df_filtered = df[df.index.isin(file_images)]

df_filtered.describe()
```

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df_filtered, test_size=0.3)

df_train.describe()
```

```python
for index in df_train.index:
  shutil.copy(f'images/{index}', f'train/{index}')
```

```python
for index in df_test.index:
  shutil.copy(f'images/{index}', f'test/{index}')
```
